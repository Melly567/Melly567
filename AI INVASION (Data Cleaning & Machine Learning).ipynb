```python
df= df.drop(['Unnamed: 0','loc',],axis=1)
df.columns
```

```python
##fill up missing values
mean_bed=df.bedroom.mean()
mean_bat=df.bathroom.mean()
mean_park=df.parking_space.mean()

df.title.fillna('Duplex',inplace=True)
df.bedroom.fillna(mean_bed,inplace=True)
df.bathroom.fillna(mean_bat,inplace=True)
df.parking_space.fillna(mean_park,inplace=True)
```

```python
df=df.drop(['title'],axis=1)
```

```python
import sklearn 
```

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df['price']#target
X=df.drop('price', axis=1)
```

```python
X_train,x_test,y_train,y_test= train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
##we are using mean_absolute_error because we are dealing with regression model 
y_predict=reg.predict(x_test)
print('Mean_absolute_error'),mean_absolute_error(y_test,y_predict)
```

```python
#finding R^2 for the model 
r2_score = reg.score(x_test,y_test)
print('the score for R^2 is :',r2_score)
print('==')
print('predicted target is ',y_predict)
```
